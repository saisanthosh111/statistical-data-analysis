## Task 2 - Simulating T Distributions

```python
x = np.linspace(ss.norm.ppf(0.01),
                ss.norm.ppf(0.99), 100)

st_vals = np.array( [ss.norm.rvs(size=1000) for _ in range(30)] )
```

```python
chi_sq_distributions = []
degress_of_freedom = [3,5,10,15,20,25,30]
for degree in degress_of_freedom:
    sq_st_vals = np.array([np.square(st_vals[i]) for i in range(degree)])
    sum_of_squares = np.sum(sq_st_vals,axis=0)
    chi_sq_distributions.append((sum_of_squares,degree))
    y = np.linspace(ss.chi2.ppf(0.01,df=degree),ss.chi2.ppf(0.99,df=degree), 100)
```

```python
for chi_distrib in chi_sq_distributions:
    z = ss.norm.rvs(size=1000)
    degree = chi_distrib[1]
    t_val = [ math.sqrt(degree)*z[i]/math.sqrt(chi_distrib[0][i]) for i in range(chi_distrib[0].shape[0]) ]
    y = np.linspace(ss.t.ppf(0.01,df=degree),ss.t.ppf(0.99,df=degree), 100)
    plt.plot(y, ss.t.pdf(y,df=degree),'r-', lw=5, alpha=0.6, label='t distribution  pdf')
    plt.hist(t_val,density=True,histtype='stepfilled')
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.title(f't distribution when v is {degree}')
    plt.legend()
    plt.show()
```

```python
for chi_distrib in chi_sq_distributions:
    z = ss.norm.rvs(size=1000)
    degree = chi_distrib[1]
    t_val = [ math.sqrt(degree)*z[i]/math.sqrt(chi_distrib[0][i]) for i in range(chi_distrib[0].shape[0]) ]
    y = np.linspace(ss.t.ppf(0.01,df=degree),ss.t.ppf(0.99,df=degree), 100)
    plt.plot(y, ss.t.pdf(y,df=degree),'r-', lw=5, alpha=0.6, label='t distribution  pdf')
    plt.plot(sorted(z), ss.norm.pdf(sorted(z)), 'y-', lw=5, alpha=0.6, label='z distribution pdf')
    plt.hist(t_val,density=True,histtype='stepfilled')
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.title(f't distribution vs z distribution  when v is {degree}')
    plt.legend()
    plt.show()
```

## Task 4 - Normality Checks

```python
# Multivariate data
brv = ss.multivariate_normal([5, 3, 7], [[4,-1,0],[-1,4,2],[0,2,9]])
multi_data = brv.rvs(size=100).T
```

### Univariate Normality Check


```python

# Univariate Q-Q Plots

def plot_qq(x,n,var_idx):

    x_new = np.array([(j-0.5)/n for j in range(n)])
    q_j = np.array([ss.norm.ppf(m) for m in x_new])
    sns.regplot(x=q_j,y=np.sort(x),scatter_kws={"color": "black"},line_kws={'color':'red'})
    plt.xlabel('q(j) (Theoretical Quantiles)')
    plt.ylabel('x(j) (Sample Quantiles)')
    plt.title(f'Q-Q Plot for variable {var_idx}\n(To check Univariate normality)')
    plt.show()
```

```python
for i in range(3):
    plot_qq(multi_data[i], 100,i+1)
```

### Bivariate Normality Check


```python
# Bi-Variate Normality test

plt.scatter(multi_data[0],multi_data[1])
plt.title('Scatter plot of Variables 1 and 2\n(To check bivariate normality)')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()

plt.scatter(multi_data[1],multi_data[2])
plt.title('Scatter plot of Variables 2 and 3\n(To check bivariate normality)')
plt.xlabel('Variable 2')
plt.ylabel('Variable 3')
plt.show()

plt.scatter(multi_data[0],multi_data[2])
plt.title('Scatter plot of Variables 1 and 3\n(To check bivariate normality)')
plt.xlabel('Variable 1')
plt.ylabel('Variable 3')
plt.show()
```

### Multivariate Normality Check

```python
def get_d_j(x_val,x_bar):
    return np.matmul(np.matmul(x_val-x_bar,S_inv),(x_val-x_bar).T).squeeze(0).squeeze(0)
```

```python

# Multi-Variate

x_vals = multi_data.T
S = np.array([[4,-1,0],[-1,4,2],[0,2,9]])
S_inv = np.linalg.inv(S)
x_bar = np.mean(x_vals,axis=0).reshape((1,3))
d_j = np.array([get_d_j(x_vals[i],x_bar) for i in range(100)]).reshape((1,100))
```

```python
def d_j_plot(d_j,n):
    x_new = np.array([(j-0.5)/n for j in range(n)])
    q_j = np.array([ss.chi2.ppf(m,3) for m in x_new])
    sns.regplot(x=q_j,y=d_j,scatter_kws={"color": "black"},line_kws={'color':'red'})
    plt.xlabel('q(c,3) (Theoretical Quantiles) ')
    plt.ylabel('d(j)^2 (Sample Quantiles) ')
    plt.title('Plot to check multivariate normality\n(Here 3 variables)')
    plt.show()
```

```python
d_j_plot(np.sort(d_j),100)
```
