### Normal Distribution

```python
# CLT Normal  Parameters: mu ,sigma 
mu = 20
sigma = 0.5
r_val = ss.norm.rvs(loc=mu,scale=sigma,size=10000)
mean_dist = [ np.mean(np.random.choice(r_val,40)) for _ in range(1000)]
```

```python

mean_sample_mean = np.mean(mean_dist)
pop_mean = np.mean(r_val)
mean_vs_mean = f'Expectation of Sample Means: {mean_sample_mean} \nMean of Population: {pop_mean}\n'
print(mean_vs_mean)

variance_sample_mean = np.std(mean_dist)**2
pop_variance = np.std(r_val)**2
variance_vs_variance = f'Variance of Sample Means: {variance_sample_mean} \nTheoretical Variance of Population: {sigma**2}\n'
print(variance_vs_variance)

sns.displot(mean_dist,kde=True,label=f'mu: {mu} sigma: {sigma}',stat='probability')
plt.xlabel('Mean of Sample')
plt.ylabel('Distribution')
plt.title('Distribution of Sample Means')
plt.legend()
plt.show()



cum_mean = [statistics.mean(mean_dist[:i]) for i in range(2,len(mean_dist))]
plt.plot(cum_mean,label='Cumulative mean')
plt.plot([pop_mean]*1000,label='Expected Mean')
plt.title('Cumulative mean vs Theoretical Mean')
plt.xlabel('Number of Simulations')
plt.ylabel('Mean')
plt.legend()
plt.show()

cum_var = [statistics.variance(mean_dist[:i]) for i in range(2,len(mean_dist))]
plt.plot(cum_var,label='Cumulative Variance')
plt.plot([pop_variance/40]*1000,label='Expected Variance (sigma**2/n)')
plt.title('Cumulative Variance vs Expected Variance')
plt.xlabel('Number of Simulations')
plt.ylabel('Variance')
plt.legend()
plt.show()
```

```python
def geometric_distribution(distrib,p):
    x_val = distrib.rvs(p,size=500)
    return (np.sort(x_val),distrib.pmf(np.sort(x_val),p),)
```
